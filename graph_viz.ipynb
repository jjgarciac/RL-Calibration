```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
k1 = np.maximum.accumulate(np.load('cheetah_k1_ph20_ps10_np12_ntrain30_cFalse.npy').squeeze())
k1_c = np.maximum.accumulate(np.load('cheetah_k1_ph20_ps10_np12_ntrain30_cTrue.npy').squeeze())
k3 = np.maximum.accumulate(np.load('cheetah_k3_ph20_ps10_np12_ntrain30_cFalse.npy').squeeze())
k3_c = np.maximum.accumulate(np.load('cheetah_k3_ph20_ps10_np12_ntrain30_cTrue.npy').squeeze())
```

```python
plt.style.use('dark_background')
fig, ax = plt.subplots()
ax.set_title('Half-Cheetah')
ax.set_ylabel('Maximum reward obtained in a horizon up to (t)')
ax.set_xlabel('Number of iterations (t)')
ax.plot(k1, label='PE')
ax.plot(k1_c, label='calibrated PE')
ax.plot(k3, label='Ours')
ax.plot(k3_c, label='calibrated Ours')
ax.legend(loc='lower right')
plt.show()
```
